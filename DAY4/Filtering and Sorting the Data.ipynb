```python
#importing the necessary packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/DELL/Desktop/DATA SCIENCE COURSE (Datasets)/Titanic (1).csv')
pd.set_option('display.max_columns',28)  # for displaying all the columns in df
```

```python
df.info()
```

```python
#Step 1: Dropping the columns having majority of values equal to 0
df.loc[df.zero == 0].shape
```

```python
#Dropping the columns having z in their names
df.drop([col for col in df.columns if 'z' in col] , axis=1,inplace=True)
```

```python
#Step 2: Renaming the columns
df.rename(columns = {'2urvived':'Survived'}, inplace = True)
```

```python
#Step 3: Filling the null values
df.Embarked.unique()
```

```python
#Maximum passengers has the value of embarked as 2.0 
df.groupby(by=df.Embarked).count()[['Passengerid']]
```

```python
df.Embarked.fillna( 2.0, inplace = True) #filled null values with 2.0
```

```python
#Step 4: Typecasting the data types
df['Age'] = df.Age.astype('int64')
df['Embarked'] = df.Embarked.astype('int64')
```

## Filtering the Data

```python
#using greater than operator to filter the dataframe
df.loc[df.Age > 30]
```

```python
#using and operator to filter dataframe based on 2 conditions
df.loc[(df.Age > 7) & (df.Age < 31)]
```

```python
#using or operator to filter dataframe based on 2 conditions
df.loc[(df.Embarked == 1) | (df.Embarked == 2)]
```

```python
#using not equal to operator to filter dataframe
df.loc[(df.Pclass !=3)]
```

## Sampling the Data

```python
#Sampling 100 rows from the dataframe
df.sample(n = 100 , random_state = 101)
```

```python
#sampling 10% rows from the dataframe
df.sample(frac=0.1 , random_state = 100)
```

## Sorting the Data

```python
#sorting the dataframe by Fare in ascending order
df.sort_values(by=['Fare'], ascending=True, inplace=False)
```

```python
#sorting the dataframe by Age in descending order
df.sort_values(by=['Age'], ascending=False, inplace=False)
```
